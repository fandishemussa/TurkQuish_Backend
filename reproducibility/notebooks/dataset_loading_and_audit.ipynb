```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import os

plt.rcParams.update({"font.size": 11, "figure.dpi": 130})
SEED = 42
os.makedirs("multiclass_results", exist_ok=True)
os.makedirs("multiclass_plots", exist_ok=True)

print("\n" + "=" * 80)
print("STEP 1 — DATA LOADING & MULTI-CLASS AUDIT")
print("=" * 80)

# Load the dataset
df = pd.read_csv("dataset_with_folds_v10.csv")
print(f"\n[1] Loaded: {len(df):,} rows")
print(f"    Columns: {list(df.columns)}")

# Display class distribution
print(f"\n[2] Multi-class distribution (class_final):")
class_dist = df["class_final"].value_counts().sort_index()
for cls, count in class_dist.items():
    pct = 100 * count / len(df)
    print(f"    {cls:20s}: {count:7,}  ({pct:5.2f}%)")

print(f"\n[3] Binary distribution (label):")
label_dist = df["label"].value_counts()
for lbl, count in label_dist.items():
    pct = 100 * count / len(df)
    print(f"    {lbl:20s}: {count:7,}  ({pct:5.2f}%)")

# Check encoding
print(f"\n[4] Encoding verification:")
print(f"    class_enc values: {sorted(df['class_enc'].unique())}")
print(f"    label_enc values: {sorted(df['label_enc'].unique())}")

# Verify fold distribution
print(f"\n[5] Fold distribution:")
for fold in sorted(df["fold"].unique()):
    fold_data = df[df["fold"] == fold]
    print(f"    Fold {fold}: {len(fold_data):7,} rows")
    for cls in sorted(df["class_final"].unique()):
        count = len(fold_data[fold_data["class_final"] == cls])
        pct = 100 * count / len(fold_data)
        print(f"        {cls:20s}: {count:6,}  ({pct:5.2f}%)")
```
